```python
import random
x = 10
print('You start with 10 marbles, your goal is to receive twenty in total.')
while x < 20 and x > 0:
    y = random.randint(1,6)
    val = input('Enter your guess of even or odd: ')
    if val.lower() == 'even':
        if y % 2 == 0:
            x += y
            print('The value was {}. You now have {} marbles.'.format(y,x))
        elif y % 2 != 0:
            x -= y
            print('The value was {}. You now have {} marbles.'.format(y,x))
    elif val.lower() == 'odd':
        if y % 2 == 0:
            x -= y
            print('The value was {}. You now have {} marbles.'.format(y,x))
        elif y % 2 != 0:
            x += y
            print('The value was {}. You now have {} marbles.'.format(y,x))
if x >= 20:
    print('Congratulations on winning the fourth game!')
elif x < 1:
    print('You have been eliminated.')
```

```python
import random
x = 100
y = 0
q = 100
limit = 7
round = 0
print('You must get to the end without getting caught. You have six rounds')
while y < x and round < limit:
    w = random.randint(10,30)
    print('Mugunghwa Kkochi Piotsseumnida!')
    z = int(input('Enter the distance you want to run: '))
    if z > w:
        print('You have been eliminated.')
        break
    elif z < w:
        y += z
        q -= z
        round += 1
        print('You are now {} meters forward. {} meters to go.'.format(y, q))

if y >= x:
    print('Congratulations on winning the first game!')
elif round == limit:
    print('You have reached the round limit. You have been eliminated.')
```
